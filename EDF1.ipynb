```python
# considering deadline = time period
class task:
    active = False
    left_over = 0.0
    ttd = 0.0
    prev_task = 1.0 #1 means not previous task, 0 means it was the previous task
    def __init__(self, name, offset, exec_time, deadline, tp):
        self.name = name
        self.offset = offset
        self.exec_time = exec_time
        self.deadline = deadline
        self.tp = tp
    
    def assign_prior(self, x):
        self.priority = x
```

```python
T2 = task('2',450.0, 69.0, 660.0, 660.0) #p=660, c=69, pri=1, offset=450, f=15.15
T5 = task('5',373.0, 379.0, 620.0, 620.0) #p=620, c=379, pri=2, offset=373, f=16.13
T4 = task('4',386.0, 39.0, 440.0, 440.0) #p=440, c=39, pri=3, offset=386, f=22.73
T3 = task('3',161.0, 5.0, 230.0, 230.0) #p=230, c=5, pri=4, offset=161, f=43.48
T1 = task('1',48.0, 17.0, 200.0, 200.0) #p=200, c=17, pri=5, offset=48, f=50.00
```

```python
task_list = [T1,T2,T3,T4,T5]
```

```python
output = []
for i in range(10000):
    #print(i)
    active_tasks = []
    current_task = ''
    for j in task_list:
        if (i>=j.offset) and ((i-j.offset)%(j.tp) == 0):
            j.active = True
            j.left_over = j.exec_time
            j.ttd = j.deadline
        elif j.left_over > 0:
            j.active = True
        else:
            j.active = False
            j.ttd = 0.0
        #print(j.name,j.active)
        
    dict1 = {T1:T1.ttd, T2:T2.ttd, T3:T3.ttd, T4:T4.ttd, T5:T5.ttd}
    #above order makes a difference during ties among 2 or more tasks, when none of them were the previous task.
    #print({k.name:k.ttd for k in dict1},{k.name:k.prev_task for k in dict1})
    list1 = sorted(dict1.items(), key=lambda item: (item[1], item[0].prev_task))
    
    for q in range(1, len(list1)+1):
        list1[q-1][0].assign_prior(q)
    
    task_list = sorted(task_list, key=lambda task: task.priority)
    
    for task in task_list:
        task.prev_task=1.0
        if task.active==True:
            active_tasks.append(task)
    #print([k.name for k in active_tasks])
    if len(active_tasks)>0:
        current_task = active_tasks[0].name
        active_tasks[0].left_over-=1
        active_tasks[0].prev_task=0.0
        for AT in active_tasks:
            AT.ttd-=1
    if len(active_tasks)==0:
        output.append(0)
    else:
        output.append(int(current_task))
    #print("time:",i,", task executed:",current_task, ", new time to deadline:",{k.name:k.ttd for k in active_tasks})
```
